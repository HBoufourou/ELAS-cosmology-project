```python
!pip install pretty_midi
!pip install note-seq

import pretty_midi
import note_seq

# Création d'une séquence MIDI simple, tu peux modifier les notes/patterns !
midi = pretty_midi.PrettyMIDI()
piano = pretty_midi.Instrument(program=0)

notes = [60, 62, 64, 65, 67, 69, 71, 72]  # Do Ré Mi Fa Sol La Si Do
start = 0.0
for note_number in notes:
    note = pretty_midi.Note(velocity=100, pitch=note_number, start=start, end=start+0.3)
    piano.notes.append(note)
    start += 0.3

midi.instruments.append(piano)
midi.write('test_melodie.mid')

# Lecture dans Colab avec note_seq
seq = note_seq.midi_file_to_note_sequence('test_melodie.mid')
note_seq.play_sequence(seq, synth=note_seq.fluidsynth)

# Téléchargement du fichier MIDI
from google.colab import files
files.download('test_melodie.mid')
```
