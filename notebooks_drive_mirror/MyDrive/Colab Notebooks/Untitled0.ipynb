```python
import magenta.music as mm
import note_seq
import pretty_midi
from note_seq.protobuf import music_pb2

# Fonction utilitaire pour créer une pattern rythmique simple
def create_drum_track(pattern, velocity=80):
    seq = music_pb2.NoteSequence()
    for i, hit in enumerate(pattern):
        if hit:
            # Instrument 9 = Drums
            seq.notes.add(pitch=36, start_time=i*0.25, end_time=i*0.25+0.1, velocity=velocity, is_drum=True)
    seq.total_time = len(pattern) * 0.25
    return seq

# Rythme inspiré du chaabi marocain (8 temps)
chaabi_rhythm = [1, 0, 0, 1, 1, 0, 1, 0]
# Rythme gnawa (cloche/krakeb)
gnawa_rhythm = [1, 0, 1, 0, 1, 1, 0, 1]
# Layering les deux patterns
drums_chaabi = create_drum_track(chaabi_rhythm, 100)
drums_gnawa = create_drum_track(gnawa_rhythm, 80)

# Fusionner les deux tracks percussives
drums = mm.sequences_lib.concatenate_sequences([drums_chaabi, drums_gnawa], sequence_durations=[2, 2])

# Basse simple (comme le guembri en gnawa)
bass_seq = music_pb2.NoteSequence()
notes_bass = [43, 43, 46, 48, 43, 46, 48, 43]  # Notes inspirées du gnawa (mi, sol, la)
for i, pitch in enumerate(notes_bass):
    bass_seq.notes.add(pitch=pitch, start_time=i*0.25, end_time=i*0.25+0.2, velocity=80)
bass_seq.total_time = 2

# Mélodie (inspirée du nay ou du chant)
melody_seq = music_pb2.NoteSequence()
melody_notes = [67, 67, 70, 72, 74, 72, 70, 67]  # (sol, la, si, do)
for i, pitch in enumerate(melody_notes):
    melody_seq.notes.add(pitch=pitch, start_time=i*0.25, end_time=i*0.25+0.2, velocity=110)
melody_seq.total_time = 2

# Superposer les couches (merge)
final_seq = mm.sequences_lib.concatenate_sequences([drums, bass_seq, melody_seq], sequence_durations=[2, 2, 2])

# Écoute directement dans Colab
note_seq.sequence_proto_to_midi_file(final_seq, 'maroc_energie.mid')
note_seq.play_sequence(final_seq, synth=note_seq.fluidsynth)

# Si tu veux télécharger le MIDI
from google.colab import files
files.download('maroc_energie.mid')
```
